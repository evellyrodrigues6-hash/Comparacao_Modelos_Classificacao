```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
test_df.head()
```

```python
test_df.info()
```

```python
# Substituindo os dados nulos pela a média
train_df['Age'].fillna(train_df['Age'].mean(), inplace=True)
train_df['Fare'].fillna(train_df['Fare'].mean(), inplace=True)

test_df['Age'].fillna(test_df['Age'].mean(), inplace=True)
test_df['Fare'].fillna(test_df['Fare'].mean(), inplace=True)
```

```python
# Aplicando o map para a variavel Sex
train_df['Sex'] = train_df['Sex'].map({'male': 0, 'female': 1})
test_df['Sex'] = test_df['Sex'].map({'male': 0, 'female': 1})
```

```python
# Criando uma nova coluna com one-hot
train_df = pd.get_dummies(train_df, columns=['Embarked'], prefix='Embarke')
test_df = pd.get_dummies(test_df, columns=['Embarked'], prefix='Embarked')
```

```python
# Criando uma nova coluna e ajustando o tipo
train_df['Cabin_Known'] = train_df['Cabin'].notna().astype(int)
test_df['Cabin_Known'] = train_df['Cabin'].notna().astype(int)
```

```python
# Apagando a colunas Cabin
train_df = train_df.drop(columns=['Cabin'])
test_df = test_df.drop(columns=['Cabin'])
```

```python
# Apagando as colunas Name e Ticket
train_df = train_df.drop(columns=['Name','Ticket'])
```

```python
test_df.dtypes
```

```python
train_df.dtypes
```

```python
for column in train_df.columns:
    if train_df[column].dtype == 'bool':
        train_df[column] = train_df[column].astype(int)
train_df.dtypes
```

```python
variaveis = ['Sex', 'Cabin_Known', 'Fare', 'Pclass', 'Age','SibSp','Parch']
```

```python
for column in test_df.columns:
    if test_df[column].dtype == 'bool':
        test_df[column] = test_df[column].astype(int)
test_df.dtypes
```

```python
# Separando X e Y
X = train_df[variaveis]
y = train_df['Survived']

X_test = test_df[variaveis]
```

```python
# Padronizando os Dados
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)
```

```python
# Balanceando os Dados
smote = SMOTE(random_state=42)
X_train_scaled_balanced, Y_train_balanced = smote.fit_resample(X_train_scaled, y)
```

```python
X_train_scaled_balanced
```

```python
Y_train_balanced
```

```python
svm = SVC(kernel = 'linear', random_state = 1, C = 1.0) 

param_grid_svm = {
    'kernel': ['rbf'],
    'C': [150, 180, 200, 220, 250],
    'gamma': [0.05, 0.08, 0.1, 0.12, 0.15]
}
grid_search_svm = GridSearchCV(estimator=svm,
                           param_grid=param_grid_svm,
                           scoring='accuracy', 
                           cv=7, 
                           verbose=2, 
                           n_jobs=-1)


grid_search_svm.fit(X_train_scaled_balanced, Y_train_balanced)
```

```python
best_params = grid_search_svm.best_params_
best_svm_model = grid_search_svm.best_estimator_
```

```python
y_pred = best_svm_model.predict(X_train_scaled_balanced)
```

```python
print("Melhores hiperparâmetros encontrados:", best_params)
```

```python
relatorio = classification_report(Y_train_balanced, y_pred)
print("Relatório de Classificação com o Melhor Modelo:")
print(relatorio)
```
